```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import scipy.stats as sp
import pylab
import statsmodels.api as sm
```

```python
df1 = pd.read_csv('Desktop/diemthi2021.csv', encoding = 'UTF-8')
df2 = pd.read_excel('Desktop/danh muc quan huyen.xlsx')
df1
df2
```

```python
df1.info()
df2.info()
```

```python
# df1
# 1. Xóa cột Unamed
# 2. Sort SBD
df1.rename({'Unnamed: 0':'a'}, axis=1, inplace=True)
df1.drop(['a'], axis=1, inplace=True)
df1.sort_values(by='id_examinee', inplace=True)
df1.reset_index(drop=True, inplace=True)
df1

# df2
# 1. Set header
# 2. Xóa cột STT, Ma_Quan/Huyen, Quan/Huyen
# 3. Xóa duplicate
df2.columns = df2.iloc[0]
df2 = df2[1:]
df2.drop(['STT','Ma_Quan/Huyen','Quan_Huyen'], axis=1, inplace=True)
df2.drop_duplicates(['Ma_Tinh/TP'],inplace=True)
df2.reset_index(drop=True, inplace=True)
df2
```

```python
df2.nunique()
```

```python
# Tách Mã tỉnh - SBD
Ma_Tinh_TP = []
cache = df1['id_examinee'].tolist()
for i in cache:
    j = i // 1000000
    Ma_Tinh_TP.append(j)
df1['Ma_Tinh/TP'] = Ma_Tinh_TP
df1
```

```python
# Merge 2 dataframes
df = pd.merge(df1, df2, how='inner', on='Ma_Tinh/TP')
df
```

Học sinh ở địa phương kém phát triển được cộng điểm thi THPT Quốc gia là phù hợp? (+1đ) (Mức ý nghĩa 5%)
1. H0: Học sinh nông thôn có điều kiện học tập kém được +1đ sẽ ngang bằng với học sinh thành thị có điều kiện học tập tốt.
                 μ1 = μ2' (μ2' = μ2 + 1)  
2. H1: Có sự khác biệt lớn giữa điểm thi của học sinh thành thị với điểm thi của học sinh nông thôn được + 1đ.
                 μ1 ≠ μ2'

Dataset sẽ được chia thành 2 phần
1. Điểm thi của những thí sinh ở Hà Nội và Tp. Hồ Chí Minh (n1), lấy ngẫu nhiên 10000 giá trị
2. Điểm thi của những thí sinh ở Hà Giang, Quảng Nam, Kon Tum và Sóc Trăng (n2), lấy ngẫu nhiên 10000 giá trị

```python
df_n1 = df.apply(lambda row: row[df['Tinh/TP'].isin(['Hà Nội', 'Tp. Hồ Chí Minh'])])
df_n1
df_n1.info()

df_n2 = df.apply(lambda row: row[df['Tinh/TP'].isin(['Hà Giang', 'Kon Tum', 'Quảng Nam', 'Sóc Trăng'])])
df_n2
df_n2['Tinh/TP'].unique()
df_n2.info()
```

MATHEMATICS

```python
# Nhóm 1
n1_math = df_n1['math'].dropna().sample(10000)
n1_math.describe()
```

```python
# Nhóm 2
n2_math = df_n2['math'].dropna().sample(10000)
n2_math.describe()
n2_math_new = (df_n2['math'].dropna() +1).sample(10000)
n2_math_new.describe()
```

```python
score_math = {'n1': list(n1_math), 'n2': list(n2_math)}
df_score = pd.DataFrame(score_math)
sns.boxplot(data=df_score)
plt.show()

score_math_new = {'n1': list(n1_math), 'n2': list(n2_math_new)}
df_score_new = pd.DataFrame(score_math_new)
sns.boxplot(data=df_score_new)
plt.show()
```

```python
print(np.var(n1_math), np.var(n2_math), np.var(n2_math_new))
```

```python
stat, p = stats.levene(n1_math, n2_math)
stat, p
stat, p = stats.levene(n1_math, n2_math_new)
stat, p
```

```python
stats.ttest_ind(n1_math, n2_math, equal_var=False)
stats.ttest_ind(n1_math, n2_math_new, equal_var=False)
```
